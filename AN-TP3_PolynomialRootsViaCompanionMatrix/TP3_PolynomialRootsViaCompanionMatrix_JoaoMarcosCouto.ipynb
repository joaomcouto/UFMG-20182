```python
import numpy as np 
```

```python


def RaizesPolinomioViaMatrizCompanheira(coef):
    n = (len(coef)-1)
    comp = np.zeros((n,n))
    i = 0
    k = 1
    for j in coef [1:] :
        comp[0,i] = -(j/coef[0])
        if (i < n-1):
            comp[k,i] = 1
            i+= 1
            k+=1
    #print (comp) 
    roots = np.linalg.eigvals(comp)
    
    return roots    
```

```python

coefP1= np.array([4, 20, 25, -10, -20 , 8])
coefP2= np.array([231/16, 0, -315/16, 0,  105/16, 0, -5/16])
coefP3= np.array([1/24, -5/6, 5, -10, 5])
coefP4= np.array([32,0, -160,0, 120,0])

raizP1 = RaizesPolinomioViaMatrizCompanheira(coefP1)
print ("As raizes de P1 são: " , raizP1)
raizP2 = RaizesPolinomioViaMatrizCompanheira(coefP2)
print ("As raizes de P2 são: " , raizP2)
raizP3 =  RaizesPolinomioViaMatrizCompanheira(coefP3)
print ("As raizes de P3 são: " , raizP3)
raizP4 = RaizesPolinomioViaMatrizCompanheira(coefP4)
print ("As raizes de P4 são: " , raizP4)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.subplots(1,1,figsize=(10,4))
real_roots = raizP1.real
xmin = min(real_roots)
xmax = max(real_roots)
x = np.linspace(xmin,xmax)
y = np.polyval(coefP1, x)
plt.plot(x,y, color = 'black')
plt.plot(real_roots,[0]*len(real_roots),'d', color = 'green')
```
